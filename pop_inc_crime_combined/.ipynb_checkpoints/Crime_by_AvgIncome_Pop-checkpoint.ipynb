```python
# Import Dependencies
import pandas as pd
```

```python
# Create a reference the CSV file desired
csv_path = "Resources/crime_grouped.csv"

# Read the CSV into a Pandas DataFrame
crime_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
crime_df.head()
```

```python
# Create a reference the CSV file desired
csv_path = "Resources/crime_incbyregion.csv"

# Read the CSV into a Pandas DataFrame
inc_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
inc_df.head()
```

```python
# Create a reference the CSV file desired
csv_path = "Resources/crime_popbyregion.csv"

# Read the CSV into a Pandas DataFrame
pop_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
pop_df.head()
```

```python
#merge tables using the variable Region and Year
combined_df = pd.merge(crime_df, inc_df, on=['Region', 'Year'])
combined_df.head()
```

```python
#merge tables again using the variable Region and Year
combined_1_df = pd.merge(combined_df, pop_df, on=['Region', 'Year'])
combined_1_df.head()
```

```python
# Export file as a CSV
combined_1_df.to_csv("Resources/combined.csv", index=False, header=True)
```
